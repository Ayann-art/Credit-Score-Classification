#cleaning

```python
import pandas as pd
df=pd.read_csv('/content/credit-score-classification/train.csv')
```

```python
df
```

```python
df.describe()
```

```python
print(f"Missing\n{df.isnull().sum()}")  #bos deyerler
```

```python
df = df.dropna(axis=0)  #bos olan nan olan setirleri silir
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.shape
df.info()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # sÉ™hvli dÉ™yÉ™rlÉ™ri NaN edÉ™cÉ™k
df['Age'] = df['Age'].astype('Int64')  # int tipinÉ™ Ã§evirir, NaN-larÄ± saxlayÄ±r

# NÉ™ticÉ™ni yoxlamaq
print(df.dtypes)
print(df['Age'].head())
```

```python
Q1 = df['Monthly_Inhand_Salary'].quantile(0.25)
Q3 = df['Monthly_Inhand_Salary'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Monthly_Inhand_Salary'] >= Q1 - 1.5*IQR) &
        (df['Monthly_Inhand_Salary'] <= Q3 + 1.5*IQR)]
df
```

```python
df['Age'] = df['Age'].astype(str)
df['Age'] = df['Age'].str.replace(r'\D', '', regex=True)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce').astype('Int64')
df
```

```python
df['Monthly_Inhand_Salary'] = df['Monthly_Inhand_Salary'].round(0).astype('Int64')
df
```

```python
df.shape
df.info()
```

```python
df['ID'] = df['ID'].str.upper()
print(df['ID'].head())
```

```python
# SÃ¼tunu É™vvÉ™lcÉ™ string tipinÉ™ Ã§evir
df['Annual_Income'] = df['Annual_Income'].astype(str)

# SonundakÄ± "_" simvolunu silmÉ™k
df['Annual_Income'] = df['Annual_Income'].str.replace(r'_+$', '', regex=True)

# NÉ™ticÉ™ni yoxlamaq
print(df['Annual_Income'].head())
```

```python
df
```

```python
# Ä°cazÉ™ verilÉ™n dÉ™yÉ™rlÉ™r
allowed_values = [
    'High_spent_Small_value_payments',
    'Low_spent_Small_value_payments'
]

# ÅžÉ™rtÉ™ uyÄŸun olmayanlarÄ± None etmÉ™k
df.loc[~df['Payment_Behaviour'].isin(allowed_values), 'Payment_Behaviour'] = None
df
```

```python
df['Name'] = df['Name'].str.split().str[0]
```

```python
df
```

```python
df['Customer_ID'] = df['Customer_ID'].str.lower()
df
```

```python
df['Amount_invested_monthly'] = pd.to_numeric(
    df['Amount_invested_monthly'], errors='coerce'
).round(0).astype('Int64')
```

```python
df
```

```python
df['Total_EMI_per_month'] = pd.to_numeric(
    df['Total_EMI_per_month'], errors='coerce'
).map('{:.2f}%'.format)
df
```
